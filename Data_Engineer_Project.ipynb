```python
import psycopg2
import pandas as pd
```

```python
def create_database():
    
    # to make connection with database
    conn = psycopg2.connect("host=localhost user=postgres dbname=postgres password=postgres")
    conn.set_session(autocommit=True)
    
    # to create cursor for queries execution
    cur = conn.cursor()
    
    # run query
    cur.execute("DROP DATABASE IF EXISTS accounts")
    cur.execute("CREATE DATABASE accounts")
    
    # for closing current connection
    conn.close()
    
    conn = psycopg2.connect("host=localhost user=postgres dbname=accounts password=postgres")
    cur = conn.cursor()
    
    return cur, conn
```

```python
# store the multilines string query in variable 
accountscountry_table_create = ("""CREATE TABLE IF NOT EXISTS accountscountry(Country_Code VARCHAR PRIMARY KEY,
Long_Name VARCHAR,
Income_Group VARCHAR,
Region VARCHAR,
Lending_category VARCHAR)""")
# run the query
cur.execute(accountscountry_table_create)
conn.commit()
```
